```python
import tifffile as tiff
import numpy as np
import matplotlib.pyplot as plt

# Ruta del archivo .tif que deseas reprocesar
tif_path = '/home/enrique/Desktop/VARIOS/TFM/tfm-identificacion-troncos/data/P28/1cm_maxint/P28_16.tif'

# Leer la imagen .tif
image = tiff.imread(tif_path)

# Mostrar el rango de valores antes de la corrección para referencia
print(f"Rango de valores de la imagen antes de la corrección: min={np.min(image)}, max={np.max(image)}")

# Reemplazar los valores que están cerca de -3.4e+38 (por ejemplo, todos los valores menores a -1e+38)
nodata_value = -3.4e+38
image = np.where(image < -1e+38, 250, image)

# Mostrar el rango de valores después de la corrección
print(f"Rango de valores de la imagen después de la corrección: min={np.min(image)}, max={np.max(image)}")

# Mostrar la imagen procesada
plt.figure(figsize=(12, 12), dpi=150)
plt.imshow(image, cmap='gray')
plt.title('Imagen reprocesada con máscara de NoData')
plt.axis('off')
plt.show()
```
